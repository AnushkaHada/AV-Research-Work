```python
import csv
import re

# Read the saved log
with open('rewards_log.txt', 'r') as f:
    lines = f.readlines()

# Extract episode and reward
data = []
for line in lines:
    match = re.match(r"Episode (\d+) reward: ([\-\d.]+)", line)
    if match:
        episode = int(match.group(1))
        reward = float(match.group(2))
        data.append((episode, reward))

# Save to CSV
with open('rewards.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Episode', 'Reward'])
    writer.writerows(data)

print(f"Saved {len(data)} rewards to rewards.csv")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV
df = pd.read_csv('rewards.csv')

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Reward'], label='Episode Reward')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('PPO Rewards per Episode')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
