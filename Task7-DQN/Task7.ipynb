```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd

# read the file
with open('reward_log.txt', 'r') as f:
  lines = f.readlines()

rewards = []
epsilons = []
episodes = []

for line in lines:
    match = re.match(r"Episode (\d+) - Reward: ([\-\d.]+) - Epsilon: ([\d.]+)", line.strip())
    if match:
        ep, rew, eps = match.groups()
        episodes.append(int(ep))
        rewards.append(float(rew))
        epsilons.append(float(eps))

# Calculate moving average
window = 5
moving_avg = pd.Series(rewards).rolling(window).mean()
```

```python
# Plot 1: Episode Reward Over Time
plt.figure(figsize=(12, 6))
plt.plot(episodes, rewards, label="Reward", marker='o')
plt.plot(episodes, moving_avg, label=f"{window}-Ep Moving Avg", linestyle="--")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Episode Reward Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Epsilon Decay
plt.figure(figsize=(10, 4))
plt.plot(episodes, epsilons, label="Epsilon", color='orange')
plt.xlabel("Episode")
plt.ylabel("Epsilon")
plt.title("Epsilon Decay Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Reward vs Epsilon
plt.figure(figsize=(8, 6))
plt.scatter(epsilons, rewards, alpha=0.7)
plt.xlabel("Epsilon")
plt.ylabel("Reward")
plt.title("Reward vs Epsilon")
plt.grid(True)
plt.tight_layout()
plt.show()
```
