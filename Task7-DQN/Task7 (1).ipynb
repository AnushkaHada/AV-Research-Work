```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd

# read the file
with open('reward_log.txt', 'r') as f:
  lines = f.readlines()

rewards = []
epsilons = []
episodes = []

for line in lines:
    match = re.match(r"Episode (\d+) - Reward: ([\-\d.]+) - Epsilon: ([\d.]+)", line.strip())
    if match:
        ep, rew, eps = match.groups()
        episodes.append(int(ep))
        rewards.append(float(rew))
        epsilons.append(float(eps))

# Calculate moving average
window = 5
reward_series = pd.Series(rewards)
moving_avg = reward_series.rolling(window=window, min_periods=1).mean()
```

```python
# Plot 1: Episode Reward Over Time
plt.figure(figsize=(12, 6))
plt.plot(episodes, rewards, label="Reward", marker='o')
plt.plot(episodes, moving_avg, label=f"{window}-Ep Moving Avg", linestyle="--")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Episode Reward Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
